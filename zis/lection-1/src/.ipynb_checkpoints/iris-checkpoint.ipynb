```python
import numpy as np
data = np.genfromtxt('../data/iris.data', delimiter=',')
```

```python
sepal_length = data[:,0]
sepal_width = data[:,1]
petal_length = data[:,2]
petal_width = data[:,3]

print(sepal_length)
print(sepal_width)
print(petal_length)
print(petal_width)
```

```python
!pip install matplotlib
import matplotlib as mpl
import matplotlib.pyplot as plt

# построение графика для выявления зависимости типа цветка ириса от длины и ширины наружной доли околоцветника
plt.figure(1)
setosa = plt.plot(sepal_length[:50], sepal_width[:50], 'ro', label='Setosa')
versicolor = plt.plot(sepal_length[50:100], sepal_width[50:100], 'g^', label='Versicolor')
verginica = plt.plot(sepal_length[100:], sepal_width[100:], 'bs', label='Verginica')
plt.legend(bbox_to_anchor=(1.05, 1), loc=2, borderaxespad=0.)

plt.xlabel('Sepal Length')
plt.ylabel('Sepal Width')
plt.show()
```

```python
# построение графика для выявления зависимости типа цветка ириса от длины и ширины внутренней доли околоцветника
plt.figure(1)
setosa = plt.plot(petal_length[:50], petal_width[:50], 'ro', label='Setosa')
versicolor = plt.plot(petal_length[50:100], petal_width[50:100], 'g^', label='Versicolor')
verginica = plt.plot(petal_length[100:], petal_width[100:], 'bs', label='Verginica')
plt.legend(bbox_to_anchor=(1.05, 1), loc=2, borderaxespad=0.)

plt.xlabel('Petal Length')
plt.ylabel('Petal Width')
plt.show()
```

```python
# построение графика для выявления зависимости типа цветка ириса от длины наружной и внутренней доли околоцветника
plt.figure(1)
setosa = plt.plot(sepal_length[:50], petal_length[:50], 'ro', label='Setosa')
versicolor = plt.plot(sepal_length[50:100], petal_length[50:100], 'g^', label='Versicolor')
verginica = plt.plot(sepal_length[100:], petal_length[100:], 'bs', label='Verginica')
plt.legend(bbox_to_anchor=(1.05, 1), loc=2, borderaxespad=0.)

plt.xlabel('Sepal Length')
plt.ylabel('Petal Length')
plt.show()
```
